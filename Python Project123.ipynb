```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
pd.set_option("display.max_columns",24)
```

```python
df = pd.read_excel(r"D:\Spandana\Excellenc\Superstore_USA.xlsx")
```

```python
df.head(2)
```

```python
df["Order Year"]=df["Order Date"].dt.year
```

```python
df.info()
```

```python
df["Order Year"].value_counts()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Order Year",data=df)
plt.title("Count of Order Year")
plt.savefig("Count of Order Year.jpg")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x="Product Category", y="Profit", data=df, estimator=np.sum)
plt.title("Profit wrt Product Category")
plt.savefig("Profit wrt Product Category.jpg")
plt.show()
```

```python
df["State or Province"].value_counts()[:5]
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x="Product Category", y="Product Base Margin", data=df, estimator=np.sum)
plt.title("Product Base margin wrt Product Category")
plt.savefig("Product Base Margin wrt Product category.jpg")
plt.show()
```

```python
df['Product Base Margin'] = df['Product Base Margin'].fillna(df['Product Base Margin'].mean())

```

#Customer Segnment

#Shipping Mode

```python
df['Ship Mode'].value_counts()
```

```python
x=df['Ship Mode'].value_counts().index
y=df['Ship Mode'].value_counts().values
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x="Ship Mode",data=df,hue="Product Category")
plt.title("Product Category Graph")
plt.savefig("Product Category Graph.jpg")
plt.show()
```
